```python
#LINEAR REGRESSION
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create a simple dataset
data = {
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7],
    'Salary': [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189]
}
df = pd.DataFrame(data)

print(df.head())
# Independent variable (feature)
# Note the double brackets to keep it as a DataFrame
X = df[['YearsExperience']]

# Dependent variable (target)
y = df['Salary']
# 1. Create an instance of the model
model = LinearRegression()

# 2. Train the model using our data
# The .fit() method finds the optimal slope (m) and intercept (b)
model.fit(X, y)

print("Model training complete! âœ…")
# Get the intercept (b)
print(f"Intercept (b): {model.intercept_}")

# Get the slope (m)
print(f"Slope/Coefficient (m): {model.coef_[0]}")
# We need to pass a 2D array for prediction
years = np.array([[5]])
predicted_salary = model.predict(years)

print(f"Predicted salary for 5 years experience: ${predicted_salary[0]:,.2f}")
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.title('Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#Multiple Linear Regression
# Create a dataset with multiple features
data_multi = {
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7],
    'Age': [22, 23, 24, 25, 26, 28, 29, 30, 31, 33],
    'Salary': [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189]
}
df_multi = pd.DataFrame(data_multi)

# X now includes both features
X_multi = df_multi[['YearsExperience', 'Age']]
y_multi = df_multi['Salary']

# Create and train a new model
multi_model = LinearRegression()
multi_model.fit(X_multi, y_multi)

# Print coefficients for each feature
print(f"Intercept: {multi_model.intercept_}")
print(f"Coefficients: {multi_model.coef_}")
# The output will be an array with two values: one for YearsExperience and one for Age.

# Predict for someone with 5 years experience and age 35
prediction_data = np.array([[5, 35]])
predicted_salary_multi = multi_model.predict(prediction_data)
print(f"Predicted salary: ${predicted_salary_multi[0]:,.2f}")
```

```python
# Categorical Naive Bayes
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB

# 1. Get the Data
data = {
    'Outlook': ['Sunny', 'Rainy', 'Sunny', 'Overcast', 'Rainy', 'Sunny'],
    'Temperature': ['Hot', 'Mild', 'Hot', 'Cool', 'Mild', 'Mild'],
    'Windy': ['False', 'True', 'True', 'False', 'False', 'True'],
    'Play': ['No', 'No', 'No', 'Yes', 'Yes', 'Yes'] # This is our target
}
df = pd.DataFrame(data)

# 2. Prepare the Data
# Separate features (X) and target (y)
X = df[['Outlook', 'Temperature', 'Windy']]
y = df['Play']

# Scikit-learn models need numeric input, so we encode our categories
encoder = OrdinalEncoder()
X_encoded = encoder.fit_transform(X)

# 3. Choose and Train the Model
model = CategoricalNB()
model.fit(X_encoded, y)

# 4. Make a Prediction
# Let's predict for a new day: Outlook='Sunny', Temperature='Cool', Windy='False'
# First, we must encode this new data using the SAME encoder
new_data = encoder.transform([['Sunny', 'Cool', 'False']])
prediction = model.predict(new_data)

print(f"Prediction for ['Sunny', 'Cool', 'False']: Play = {prediction[0]}")
# You can see the probability for each class
print(f"Probabilities: {model.predict_proba(new_data)}")
```

```python
#Gaussian Naive Bayes
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB

# 1. Get the Data
data = {
    'Length': [6.0, 5.5, 7.0, 5.0, 6.5, 5.8],
    'Width': [3.0, 2.5, 3.5, 2.0, 3.2, 2.8],
    'Fruit': ['Apple', 'Grape', 'Apple', 'Grape', 'Apple', 'Grape'] # Target
}
df = pd.DataFrame(data)

# 2. Prepare the Data (No encoding needed for features!)
X = df[['Length', 'Width']]
y = df['Fruit']

# 3. Choose and Train the Model
model = GaussianNB()
model.fit(X, y)

# 4. Make a Prediction
# Predict for a new fruit with Length=6.2 and Width=2.9
new_fruit = np.array([[6.2, 2.9]])
prediction = model.predict(new_fruit)

print(f"Prediction for Length=6.2, Width=2.9: Fruit = {prediction[0]}")
```

```python
#Multinomial Naive Bayes
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Get the Data
data = {
    'message': [
        'offer sale viagra now',
        'report due tomorrow',
        'free money now',
        'meeting at 5pm',
        'urgent viagra offer'
    ],
    'label': ['spam', 'not spam', 'spam', 'not spam', 'spam'] # Target
}
df = pd.DataFrame(data)

# 2. Prepare the Data (Feature Engineering for text)
X = df['message']
y = df['label']

# Convert text into a matrix of token counts
vectorizer = CountVectorizer()
X_counts = vectorizer.fit_transform(X)

# 3. Choose and Train the Model
model = MultinomialNB()
model.fit(X_counts, y)

# 4. Make a Prediction
# Predict for a new message
new_message = ["free report about money"]
new_counts = vectorizer.transform(new_message) # Use the same vectorizer
prediction = model.predict(new_counts)

print(f"Prediction for '{new_message[0]}': Label = {prediction[0]}")
```

```python
#Bernoulli Naive Bayes
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB

# 1. Get the Data (Same data as before)
data = {
    'message': [
        'offer sale viagra now',
        'report due tomorrow',
        'free money now',
        'meeting at 5pm',
        'urgent viagra offer'
    ],
    'label': ['spam', 'not spam', 'spam', 'not spam', 'spam']
}
df = pd.DataFrame(data)

# 2. Prepare the Data
X = df['message']
y = df['label']

# Convert text into a BINARY matrix (word is present=1, or not=0)
# This is the only line that changes from the Multinomial example!
vectorizer = CountVectorizer(binary=True)
X_binary = vectorizer.fit_transform(X)

# 3. Choose and Train the Model
model = BernoulliNB()
model.fit(X_binary, y)

# 4. Make a Prediction
new_message = ["free report about money"]
new_binary = vectorizer.transform(new_message)
prediction = model.predict(new_binary)

print(f"Prediction for '{new_message[0]}': Label = {prediction[0]}")
```

```python
#Logistic regression, Binary Classification
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Create a suitable dataset
data = {
    'HoursStudied': [0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0, 3.25, 3.5, 3.75, 4.0, 4.25, 4.5, 4.75, 5.0, 5.5],
    'Pass':         [0,   0,    0,   0,    0,   0,    1,   0,    1,   0,    1,   1,    0,   1,    1,   1,    1,   1,    1,   1]
}
df = pd.DataFrame(data)
# Independent variable (feature)
X = df[['HoursStudied']]
# Dependent variable (target)
y = df['Pass']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Create an instance of the model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

print("Model training complete! âœ…")
# Make predictions on the test set
y_pred = model.predict(X_test)

# Predict the probability for the test set
y_pred_proba = model.predict_proba(X_test)

print(f"Test Data:\n{X_test.join(y_test)}")
print("\nPredictions:", y_pred)
# predict_proba returns probabilities for [class 0, class 1]
print("Prediction Probabilities:\n", y_pred_proba)

# Check the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy * 100:.2f}%")
```

```python
#Logistic regression, Multiple Classification
# Create a suitable multi-class dataset
data_multi = {
    'Alcohol': [14.2, 13.2, 13.1, 14.3, 13.2, 12.0, 12.2, 12.8, 12.5, 13.0, 11.0, 11.5, 11.8],
    'Acidity': [5.6, 3.9, 5.0, 5.9, 4.5, 3.1, 2.9, 3.5, 3.3, 3.8, 2.5, 2.2, 2.7],
    'WineType': ['A', 'A', 'A', 'A', 'B', 'C', 'C', 'B', 'B', 'B', 'C', 'C', 'C'] # Target with 3 classes
}
df_multi = pd.DataFrame(data_multi)
# Features
X_multi = df_multi[['Alcohol', 'Acidity']]
# Target
y_multi = df_multi['WineType']

# Split data
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_multi, test_size=0.3, random_state=42)
# Create an instance of the model
# It automatically handles the multi-class case
multi_model = LogisticRegression()

# Train the model
multi_model.fit(X_train_m, y_train_m)

print("Multi-class model training complete! âœ…")
# Make predictions on the test set
y_pred_m = multi_model.predict(X_test_m)

# Check the accuracy
accuracy_m = accuracy_score(y_test_m, y_pred_m)
print(f"\nTest Data:\n{X_test_m.join(y_test_m)}")
print("\nPredictions:", y_pred_m)
print(f"\nMulti-class Model Accuracy: {accuracy_m * 100:.2f}%")

# Let's predict a new wine: Alcohol=12.9, Acidity=4.1
new_wine = np.array([[12.9, 4.1]])
prediction = multi_model.predict(new_wine)
prediction_proba = multi_model.predict_proba(new_wine)

print(f"\nPrediction for new wine {new_wine}: {prediction[0]}")
print(f"Probabilities (for classes A, B, C): {prediction_proba}")
```

```python
#Perceptron
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
# Load the dataset
iris = load_iris()
# We'll use only sepal length and petal length
X = iris.data[:, [0, 2]]
y = iris.target

# Filter for only two classes: setosa (0) and versicolor (1)
X = X[y != 2]
y = y[y != 2]

print("Shape of features (X):", X.shape)
print("Shape of target (y):", y.shape)
print("Classes:", np.unique(y)) # Should be [0, 1]
# Split the data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features for better performance
scaler = StandardScaler()
# Fit the scaler ONLY on the training data to avoid data leakage
scaler.fit(X_train)

# Transform both training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
# Create an instance of the Perceptron model
# eta0 is the learning rate
# max_iter is the number of passes over the training data (epochs)
perceptron_model = Perceptron(max_iter=100, eta0=0.1, random_state=42)

# Train the model
perceptron_model.fit(X_train_scaled, y_train)

print("Perceptron model training complete! âœ…")
# Make predictions on the scaled test data
y_pred = perceptron_model.predict(X_test_scaled)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Predictions: {y_pred}")
print(f"Actual labels: {y_test}")
print(f"Model Accuracy: {accuracy * 100:.2f}%")
from matplotlib.colors import ListedColormap

def plot_decision_boundary(X, y, model, scaler):
    # Scale the features
    X_scaled = scaler.transform(X)

    # Setup plot
    plt.figure(figsize=(10, 6))

    # Define colormap
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#0000FF'])

    # Create a mesh to plot the decision boundary
    x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
    y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Make predictions on the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision regions
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot the actual data points
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=50)
    plt.title('Perceptron Decision Boundary')
    plt.xlabel('Sepal Length (Standardized)')
    plt.ylabel('Petal Length (Standardized)')
    plt.show()

# Visualize the boundary for our trained model
plot_decision_boundary(X, y, perceptron_model, scaler)
```

```python
#Multilevel Perceptron
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
# Generate a non-linear dataset with 500 samples
# noise=0.2 makes the classification task a bit more challenging
X, y = make_moons(n_samples=500, noise=0.2, random_state=42)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
plt.title("make_moons Dataset")
plt.show()
# Split the data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
scaler.fit(X_train) # Fit only on training data

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
# Create an instance of the MLPClassifier model
# hidden_layer_sizes=(10, 5) means two hidden layers:
# the first with 10 neurons, and the second with 5 neurons.
mlp_model = MLPClassifier(
    hidden_layer_sizes=(10, 5),
    activation='relu',      # ReLU is the standard activation function
    max_iter=1000,          # Number of epochs (passes over the data)
    random_state=42
)

# Train the model
mlp_model.fit(X_train_scaled, y_train)

print("MLP model training complete! âœ…")
# Make predictions on the scaled test data
y_pred = mlp_model.predict(X_test_scaled)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

print(f"Predictions: {y_pred[:10]}")
print(f"Actual labels: {y_test[:10]}")
print(f"Model Accuracy: {accuracy * 100:.2f}%")
from matplotlib.colors import ListedColormap

def plot_decision_boundary(X, y, model, scaler):
    # Setup plot
    plt.figure(figsize=(10, 6))

    # Define colormap
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])

    # Scale all data for plotting
    X_scaled = scaler.transform(X)

    # Create a mesh to plot the decision boundary
    x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
    y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Make predictions on the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision regions
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.8)

    # Plot the actual data points
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap=plt.cm.RdBu, edgecolor='k', s=50)
    plt.title('MLP Decision Boundary')
    plt.xlabel('Feature 1 (Standardized)')
    plt.ylabel('Feature 2 (Standardized)')
    plt.show()

# Visualize the boundary for our trained model using the full dataset for context
plot_decision_boundary(X, y, mlp_model, scaler)
```

```python
#SVM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix,
                             ConfusionMatrixDisplay, RocCurveDisplay)
# Generate a non-linear dataset
X, y = make_circles(n_samples=500, noise=0.1, factor=0.5, random_state=42)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualize the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu_r)
plt.title("make_circles Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler ONLY on the training data
scaler.fit(X_train)

# Transform both the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
# Define models with different kernels and parameters
models = {
    "1. Linear Kernel": SVC(kernel='linear', probability=True, random_state=42),
    "2. RBF Kernel (Default)": SVC(kernel='rbf', probability=True, random_state=42),
    "3. Poly Kernel (Degree 3)": SVC(kernel='poly', degree=3, probability=True, random_state=42),
    "4. RBF Kernel (Tuned)": SVC(kernel='rbf', C=10, gamma='auto', probability=True, random_state=42)
}

# Store results
results = []

# Loop through models to train and evaluate
for name, model in models.items():
    # Train the model
    model.fit(X_train_scaled, y_train)

    # Make predictions
    y_pred = model.predict(X_test_scaled)
    y_proba = model.predict_proba(X_test_scaled)[:, 1] # Probability for the positive class

    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)

    # Get model details
    model_num = name.split('.')[0]
    kernel_name = model.kernel
    params = f"C={model.C}, gamma={model.gamma}" if kernel_name == 'rbf' else f"degree={model.degree}" if kernel_name == 'poly' else "C=1.0"

    results.append([model_num, kernel_name, params, accuracy, precision, recall, f1, auc])

print("Model training and evaluation complete! âœ…")
# Create a DataFrame for the results
results_df = pd.DataFrame(results, columns=["Model", "Kernel", "Chosen Parameters", "Accuracy",
                                            "Precision", "Recall", "F1-Score", "AUC"])

print(results_df.to_string())
# Plotting the F1-Scores for comparison
plt.figure(figsize=(10, 6))
plt.bar(results_df['Model'] + ". " + results_df['Kernel'], results_df['F1-Score'], color=['red', 'green', 'blue', 'orange'])
plt.xlabel("Model Configuration")
plt.ylabel("F1-Score")
plt.title("Comparison of F1-Scores Across SVM Models")
plt.ylim(0, 1.1)
plt.xticks(rotation=15)
plt.show()
# Select the best model
best_model = models["4. RBF Kernel (Tuned)"]

# Generate and display the confusion matrix
cm = confusion_matrix(y_test, best_model.predict(X_test_scaled))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.classes_)
disp.plot()
plt.title("Confusion Matrix for Tuned RBF SVM")
plt.show()
# Plot the ROC curve
RocCurveDisplay.from_estimator(best_model, X_test_scaled, y_test)
plt.title("ROC Curve for Tuned RBF SVM")
plt.show()
def plot_decision_boundary(X, y, model, title):
    # Create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Predict on the meshgrid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu_r, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu_r, edgecolors='k')
    plt.title(title)
    plt.xlabel("Feature 1 (Standardized)")
    plt.ylabel("Feature 2 (Standardized)")


# Plotting boundaries
plt.figure(figsize=(16, 7))

# Plot for Linear Kernel
plt.subplot(1, 2, 1)
plot_decision_boundary(X_train_scaled, y_train, models["1. Linear Kernel"], "Linear Kernel Decision Boundary")

# Plot for Tuned RBF Kernel
plt.subplot(1, 2, 2)
plot_decision_boundary(X_train_scaled, y_train, models["4. RBF Kernel (Tuned)"], "Tuned RBF Kernel Decision Boundary")

plt.show()
```
